```python
# value_counts (serise and dataframe) 
#gives the frequency count of each uniqe item
# series
a = pd.Series([1,1,1,2,3,2,3,3,4,4,4,4,4,5])
print(a.value_counts())
marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]],columns=['iq','marks','package in lp/an'])
print(marks.value_counts())
marks
```

```python
ipl = pd.read_csv('ipl-matches.csv')
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts().plot(kind='bar')
```

```python
ipl['TossDecision'].value_counts().plot(kind='pie')
```

```python
#sort_values(ascending=True or False)
mv = pd.read_csv('movies.csv')
mv.sort_values('title_x',ascending=False)
# if there are missing values as 'NaN' then by default they will show at the 
# end of the list to get them at the top we change a hidden value
# na_position='first' --> will shoe them at the top of the table

# also we use inplace=True parameter to store the changes permanently
```

```python
# sorting in multiple levels
mv.sort_values(['year_of_release','title_x'],ascending=[True,False])
```

```python
# rank(serise)
bats = pd.read_csv('batsman_runs_ipl.csv')
bats
```

```python
bats['batsman_run'].rank(ascending=False) # here more runs means less rank
```

```python
bats['bating_rank'] = bats['batsman_run'].rank(ascending=False)
bats.sort_values('bating_rank')
```
